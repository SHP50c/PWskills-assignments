```python
# import inspect module
import inspect
# our treeClass function
def treeClass(cls, ind = 0):
    # print name of the class
    print ('-' * ind, cls.__name__)
    # iterating through subclasses
    for i in cls.__subclasses__():
        treeClass(i, ind + 3)
print("Hierarchy for Built-in exceptions is : ")
# inspect.getmro() Return a tuple
# of class clsâ€™s base classes.
# building a tree hierarchy
inspect.getclasstree(inspect.getmro(Exception))
# function call
treeClass(Exception)
```

```python
#zeroerror example
#here we are trying to devide a integer value with 0 which mathematically not
#possible hence we call exception from arithmatic error class which will invoke zero error class
try:
    1/0
except ArithmeticError as e:
    print(e)
```

```python
#overload example
#here we overload an integer value and hence the overload error is invoke from the arithmaticerror class
j = 5.0
try:
    for i in range(1, 1000):
        j = j**i
except ArithmeticError as e:
    print(e)
```
